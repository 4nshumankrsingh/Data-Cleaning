```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv(r"C:\Users\Anshuman\Downloads\train.csv")
test = pd.read_csv(r"C:\Users\Anshuman\Downloads\test.csv")
gender_submission = pd.read_csv(r"C:\Users\Anshuman\Downloads\gender_submission.csv")

print(train.head())
print(train.info())
print(train.describe())
```

```python
print(train.columns) 

if 'Cabin' in train.columns:
    train = train.drop(columns=['Cabin'])

print(train.isnull().sum())
train['Age'] = train['Age'].fillna(train['Age'].median())
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
print(train.isnull().sum())
```

```python
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1
train['IsAlone'] = np.where(train['FamilySize'] > 1, 0, 1)
train['Title'] = train['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
print(train.head())
```

```python
sns.set(style='whitegrid')

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=train)
plt.title('Distribution of Survival')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=train)
plt.title('Survival Distribution by Sex')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Pclass', data=train)
plt.title('Survival Distribution by Passenger Class')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Embarked', data=train)
plt.title('Survival Distribution by Embarked')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(data=train, x='Age', hue='Survived', kde=True, bins=30)
plt.title('Age Distribution by Survival')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='FamilySize', data=train)
plt.title('Survival Distribution by Family Size')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='IsAlone', data=train)
plt.title('Survival Distribution by Being Alone')
plt.show()

plt.figure(figsize=(14, 6))
sns.countplot(x='Survived', hue='Title', data=train)
plt.title('Survival Distribution by Title')
plt.show()
```
